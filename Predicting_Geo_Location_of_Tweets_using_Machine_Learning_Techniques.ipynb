*Sentiment analysis*

```python
#extracting 100 tweets from the twitter user account
posts=api.user_timeline(screen_name="narendramodi",count=100,lang="en",tweet_mode="extended")

#print last tweets 
print("showing last 5 tweets\n")
i=1
for tweet in posts[0:5]:
  print(str(i) + ')'+tweet.full_text+'\n')
  i=i+1
```

```python
#creating a dataframe 
df=pd.DataFrame([tweet.full_text for tweet in posts],columns=['Tweets'])
df.head()
```

```python
import re
re.compile('<title>(.*)</title>')
```

```python
from textblob import TextBlob
```

```python
#cleaning the data
#we are creting a function to clean the data 
def cleanTxt(text):
  text= re.sub(r'@[A-Za-z0-9]+','',text) #removing @
  text= re.sub(r'#','',text) #removing hashtags
  text=re.sub(r'RT[\s]+','',text) #removing RT 
  return text

  #calling the function 
df['Tweets']=df['Tweets'].apply(cleanTxt)
  #displayed cleaned data 
df
```

```python
#creating a function to get the subject
def getSubjectivity(text):
  return TextBlob(text).sentiment.subjectivity

  #to get polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
#display in two columns 
df['Subjectibity']=df['Tweets'].apply(getSubjectivity)
df['Polarity']=df['Tweets'].apply(getPolarity)

df
```

```python
#displaying in wordcloud
from wordcloud import WordCloud #This line here 
from matplotlib import pyplot as plt
allwords=' '.join([twts for twts in df['Tweets']])
wordCloud=WordCloud(width=800,height=500,random_state=21,max_font_size=120).generate(allwords)
plt.imshow(wordCloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def getAnalysis(score):
  if score<0:
    return 'Negative'
  elif score==0:
      return 'Neutral'
  else:
    return 'Positive'

df['Analysis']=df['Polarity'].apply(getAnalysis)
df
```

```python
#plot in a graph
plt.figure(figsize=(8,7))
for i in range(0,df.shape[0]):
  plt.scatter(df['Polarity'][i],df['Subjectibity'][i],color='Blue')

plt.title('Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
```

```python
# Print the percentage of negative tweets
ntweets = df[df.Analysis == 'Negative']
ntweets = ntweets['Tweets']
ntweets

round( (ntweets.shape[0] / df.shape[0]) * 100, 1)
```

```python
# Print the percentage of positive tweets
ptweets = df[df.Analysis == 'Positive']
ptweets = ptweets['Tweets']
ptweets

round( (ptweets.shape[0] / df.shape[0]) * 100 , 1)
```

```python
# Show the value counts
df['Analysis'].value_counts()
```

```python
# Plotting and visualizing the counts
plt.title('Sentiment Analysis')
plt.xlabel('Sentiment')
plt.ylabel('Counts')
df['Analysis'].value_counts().plot(kind = 'bar')
plt.show()
```
